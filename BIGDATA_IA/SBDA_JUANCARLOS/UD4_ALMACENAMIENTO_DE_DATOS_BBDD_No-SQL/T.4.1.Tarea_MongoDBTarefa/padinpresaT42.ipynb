 ### 4. Recupera aquellos pedidos (y solo el ID_Pedido) cuyo Importe total está entre 2 y 2.1 euros

```python
from pymongo import MongoClient

# Establecer conexión con MongoDB
cliente = MongoClient('172.18.0.1', 27017)
coleccion_pedidos = cliente['supermercado']['pedidos']

# Encontrar los pedidos cuyo importe total está entre 2 y 2.1 euros
pedidos_rango_importe = coleccion_pedidos.find({"Importe total": {"$gte": 2, "$lte": 2.1}}, {"ID_Pedido": 1})

# Mostrar los ID de los pedidos que cumplen con el rango de importe
print("Pedidos cuyo importe total está entre 2 y 2.1 euros:")
for pedido in pedidos_rango_importe:
    id_pedido = pedido["ID_Pedido"]
    print(f"   - ID_Pedido: {id_pedido}")
```

### 5. Pedidos que, entre sus ítems, figuran Pan y Jabón. Devuelve solo ID_Pedido

```python
pedidos_pan_jabon = coleccion_pedidos.find({
    "$and": [
        {"items": {"$elemMatch": {"producto": "Pan"}}},
        {"items": {"$elemMatch": {"producto": "Jabón"}}}
    ]
}, {"ID_Pedido": 1})

for pedido in pedidos_pan_jabon:
    print("ID_Pedido:", pedido["ID_Pedido"])
```
