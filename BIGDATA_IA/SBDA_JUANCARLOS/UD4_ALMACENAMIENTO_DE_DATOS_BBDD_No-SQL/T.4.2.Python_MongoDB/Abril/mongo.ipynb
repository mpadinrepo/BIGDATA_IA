```python
import json
from pymongo import MongoClient

# Conexión a MongoDB
client = MongoClient('mongodb://172.21.0.2:27017/')
db = client['ventas']    # creamos la base de datos
collection = db['pedidos']  # la colección

# Nombre del archivo CSV
filename = 'ventas.json'

# Insertar registros en la colección
with open(filename, 'r') as file:
    data = json.load(file)
    for row in data:
        # Insertar cada fila como un documento en la colección
        collection.insert_one(row)

print("Registros insertados exitosamente en MongoDB.")
```
