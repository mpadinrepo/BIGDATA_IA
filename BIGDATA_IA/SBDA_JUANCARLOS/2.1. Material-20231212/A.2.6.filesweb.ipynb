```python
import requests, csv
import matplotlib.pyplot as plt
path = 'C:\\Users\\Carlos.PC-PEREZABALDE\\BIGDATA\\' # or oher local folder payt attention in string syntax \\
url = "http://www.mambiente.munimadrid.es/opendata/horario.txt"
resp = requests.get(url)    # url param
with open (path + 'horariolocal.txt', 'wb') as output:
    output.write(resp.content)
with open(path + 'horariolocal.txt', 'r+') as csvfile:
    readCSV = csv.reader(csvfile, delimiter = ',')
    for row in readCSV:
        if (row[0]+row[1]+row[2]=='28079055' and row[3] == '12'):    # meteo location level and 12 NOx params
            plt.title("NIVELES DE ÓXIDO DE NITRÓGENO: " + row[8]+"/"+row[7]+"/"+row[6])
            hora = 0
            desp = 9
            vs = []
            horario = []
            while hora<=23:
                if row[desp+2*hora+1] == 'V':
                    vs.append(float(row[desp+2*hora]))
                    horario.append(hora)
                hora += 1
    plt.plot(horario, vs)
    plt.show()
```
