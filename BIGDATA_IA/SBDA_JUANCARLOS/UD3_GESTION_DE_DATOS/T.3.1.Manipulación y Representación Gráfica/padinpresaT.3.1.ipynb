## Geolocalizamos codigo postal y creamos dir_entrega

```python
import pandas as pd
import googlemaps

# Inicializar cliente de Google Maps Geocoding
gmaps = googlemaps.Client(key='TU KEY AKI')

# 1. Agregar código postal
def get_zip_code(address):
    try:
        geocode_result = gmaps.geocode(address)
        if geocode_result:
            zip_code = None
            for component in geocode_result[0]['address_components']:
                if 'postal_code' in component['types']:
                    zip_code = component['long_name']
            return zip_code
    except Exception as e:
        print(f"Error geocoding address {address}: {e}")
    return None

df_completo['Código Postal'] = df_completo['dir_entrega'].apply(get_zip_code)


# Mostrar el DataFrame resultante
print(df_completo)
```
