```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Cargar el archivo shape
archivo_shape = 'PORTAL.shp'
gdf = gpd.read_file(archivo_shape)

# Mostrar información y primeras filas de las columnas 4 a 10
columnas_interesantes = gdf.iloc[:, 3:10]
print(columnas_interesantes.head())

```

```python
gdf.columns
```

```python
# Filtrar por la calle ingresada por el usuario
valor_filtro = input("Ingrese el nombre de la calle para filtrar: ")
gdf_filtrado = gdf[gdf['NOMBRE_VIA'] == valor_filtro]

```

```python
gdf_filtrado
```

```python
# Mostrar resumen estadístico utilizando la función describe()
resumen_estadistico = gdf.describe()
resumen_estadistico

```

```python
# Visualizar todos los datos de la provincia
gdf.plot()
plt.title('Datos de la provincia de Pontevedra')
plt.show()

```

```python
# Visualizar el GeoDataFrame filtrado
gdf_filtrado.plot()
plt.title(f'Datos de la calle {valor_filtro}')
plt.show()

```

```python
# Convertir el GeoDataFrame filtrado a un DataFrame de Pandas
df_resultado = gdf_filtrado.drop(columns='geometry').copy()


```
