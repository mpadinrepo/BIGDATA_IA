```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

try:
    # Conectar con la base de datos MySQL
    conexion = mysql.connector.connect(
        host="mariadb",  # Nombre del contenedor Docker
        port="3306",  # Puerto predeterminado de MariaDB
        user="root",
        password="abc123...",  # Cambiar a la contraseña correcta
        database="mybbdd"  # Nombre de tu base de datos
    )

    # Verificar si la conexión fue exitosa
    if conexion.is_connected():
        print("Conexión establecida correctamente con la base de datos mybbdd")
        # Crear un objeto de motor SQLAlchemy
        engine = create_engine('mysql+mysqlconnector://root:abc123...@mariadb:3306/mybbdd')
        # Consulta SQL para obtener los datos de la tabla
        query = "SELECT Id, ID_Pedido, ID_Factura, Supermercado, Ciudad, Tipo_cliente, Sexo, Linea_negocio, Precio_unitario, Cantidad, Total, IVA, DATE_FORMAT(Fecha, '%Y-%m-%d') as Fecha, Pago FROM VENTAS ORDER BY (CASE WHEN Sexo = 'Mujer' THEN 1 ELSE 2 END), Sexo;"
        # Crear un DataFrame con los datos de la tabla
        datos = pd.read_sql_query(query, engine)
        # Convertir listas en cadenas concatenadas
        datos = datos.apply(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)
        
        # Convertir la columna 'Sexo' a cadenas de texto
        datos['Sexo'] = datos['Sexo'].str.decode('utf-8')  # Convertir bytes a cadenas
        
        # Exportar a archivos CSV y XLSX
        csv_file_path = 'ventas.csv'
        xlsx_file_path = 'ventas.xlsx'
        datos.to_csv(csv_file_path, index=False)
        datos.to_excel(xlsx_file_path, index=False)
        
        # Representación gráfica
        sns.countplot(x='Sexo', data=datos)
        plt.title('Distribución de Ventas por Sexo')
        plt.show()
        
        print("Archivos CSV y XLSX generados y guardados en:")
        print(f"- CSV: {csv_file_path}")
        print(f"- XLSX: {xlsx_file_path}")
        
except mysql.connector.Error as error:
    print("Error al conectar a la base de datos:", error)
finally:
    # Cerrar la conexión
    if 'conexion' in locals() and conexion.is_connected():
        conexion.close()
        print("Conexión cerrada correctamente")
```

Este script ejecutará una consulta DESCRIBE en la tabla VENTAS y mostrará la información sobre las columnas, incluyendo el nombre de la columna, el tipo de datos y otras propiedades

```python
import mysql.connector

try:
    # Conectar con la base de datos MySQL
    conexion = mysql.connector.connect(
        host="mariadb",  # Nombre del contenedor Docker
        port="3306",  # Puerto predeterminado de MariaDB
        user="root",
        password="abc123...",  # Cambiar a la contraseña correcta
        database="mybbdd"  # Nombre de tu base de datos
    )

    # Verificar si la conexión fue exitosa
    if conexion.is_connected():
        print("Conexión establecida correctamente con la base de datos mybbdd")
        
        # Consulta SQL para obtener la estructura de la tabla VENTAS
        query = "DESCRIBE VENTAS;"
        
        # Ejecutar la consulta
        cursor = conexion.cursor()
        cursor.execute(query)
        
        # Obtener los resultados
        resultados = cursor.fetchall()
        
        # Imprimir los resultados
        for columna in resultados:
            print(columna)
        
except mysql.connector.Error as error:
    print("Error al conectar a la base de datos:", error)
finally:
    # Cerrar la conexión
    if 'conexion' in locals() and conexion.is_connected():
        conexion.close()
        print("Conexión cerrada correctamente")
```

```python
import pandas as pd
import mysql.connector

def conectar_base_datos():
    try:
        # Establecer conexión con la base de datos MySQL
        conexion = mysql.connector.connect(
            host="mariadb",
            port="3306",
            user="root",
            password="abc123...",
            database="mybbdd"
        )

        # Verificar si la conexión fue exitosa
        if conexion.is_connected():
            print("Conexión establecida correctamente con la base de datos mybbdd")
            return conexion
    except mysql.connector.Error as error:
        print("Error al conectar a la base de datos:", error)
        return None

def cerrar_conexion(conexion):
    # Cerrar la conexión
    if conexion and conexion.is_connected():
        conexion.close()
        print("Conexión cerrada correctamente")

def obtener_descripciones_tablas(conexion, tablas):
    descripciones = {}
    try:
        # Iterar sobre las tablas
        for tabla in tablas:
            # Consulta SQL para obtener la descripción de la tabla
            query = f"DESCRIBE {tabla};"
            # Leer los datos en un DataFrame de pandas
            descripcion = pd.read_sql_query(query, conexion)
            # Agregar la descripción al diccionario
            descripciones[tabla] = descripcion
        return descripciones
    except mysql.connector.Error as error:
        print("Error al obtener las descripciones de las tablas:", error)
        return None

# Lista de tablas
tablas = ["categories", "customers", "employees", "orders", "products", "shippers", "suppliers", "VENTAS"]

# Ejemplo de uso
conexion = conectar_base_datos()
if conexion:
    try:
        # Obtener descripciones de tablas
        descripciones = obtener_descripciones_tablas(conexion, tablas)
        if descripciones:
            for tabla, descripcion in descripciones.items():
                print(f"Descripción de la tabla '{tabla}':")
                print(descripcion)
    finally:
        cerrar_conexion(conexion)
```
