```bash
INTRODUCCIÓN

Python es un lenguaje interpretado. Para ejecutar un programa escrito en Python necesitamos un intérprete. Algunas opciones comunes son:

CPython
PyPy
Jython
Otros que puedan surgir
Dentro de Python, la especificación se organiza por versiones, como 3.3, 3.4, ..., 3.10. Un programa escrito en una versión específica requiere al menos esa versión del intérprete para ejecutarse.

Es importante tener en cuenta que Python 2.7 ya está obsoleto, por lo que es mejor evitar su uso.



```bash
INSTALACIÓN
La instalación de Python es muy sencilla. En sistemas basados en Debian (como Ubuntu), puedes instalarlo con:
apt install python
En Windows, simplemente descarga e instala desde el sitio web oficial de Python.

PIP
PIP es el administrador de paquetes original de Python. Para instalarlo, en sistemas basados en Debian, puedes usar:

apt install python3-pip
Con PIP, puedes instalar y gestionar librerías de Python. Algunos comandos útiles son:
Instalar una librería: pip install nombre_lib
Desinstalar una librería: pip uninstall nombre_lib
Ver los paquetes instalados: pip list

ENTORNOS VIRTUALES
Los entornos virtuales son espacios aislados de Python que te permiten trabajar en proyectos con librerías y versiones específicas. Para crear un entorno virtual, puedes usar venv o conda.

Con venv, el proceso sería:

python3 -m venv /path/to/new/virtual/environment
source /path/to/new/virtual/environment/bin/activate

Con conda, sería:
conda create --name nombre_de_entorno
conda activate nombre_de_entorno
Una vez activado, los comandos pip y python se referirán a ese entorno virtual.

CONDA
Conda es una herramienta que ayuda a gestionar entornos y paquetes en Python. Puedes instalar Anaconda o Miniconda desde su sitio web oficial.

Para crear un entorno con Conda:
conda create --name nombre_del_entorno
Puedes especificar la versión de Python con:

conda create --name nombre_del_entorno python=3.6
Y para activarlo:
conda activate nombre_del_entorno

ESTRUCTURA DE UN PROYECTO
Es recomendable estructurar tus proyectos Python de manera organizada. Una estructura común incluye carpetas como bin, tests y docs. También es importante tener un archivo setup.py que describa tu proyecto y sus dependencias.

El contenido de setup.py podría ser algo así:
from setuptools import find_packages, setup

setup(
    name='NOMBRE_DEL_PROYECTO',
    packages=find_packages(),
    version='0.1.0',
    description='Descripción del proyecto',
    author='Tu Nombre',
    license='MIT',
    install_requires=['lib1', 'lib2'],
    setup_requires=['pytest-runner'],
    tests_require=['pytest'],
    test_suite='tests',
)

```bash
COMENTARIOS
Los comentarios en Python se realizan con el carácter #. Pueden estar en su propia línea o al final de una línea de código.

Por ejemplo:

variable = 1  # Este es un comentario en línea
Es importante usar comentarios para explicar el código y hacerlo más legible.

PALABRAS CLAVE
Python tiene un conjunto de palabras clave que tienen un significado específico y no pueden ser usadas como nombres de variables. Para ver la lista de palabras clave, puedes usar:
help("keywords")

TIPOS DE DATOS PREDEFINIDOS
Python tiene varios tipos de datos predefinidos, incluyendo números (enteros, flotantes, complejos), booleanos, cadenas, listas, tuplas, diccionarios y conjuntos.



```bash
OPERADORES
Python incluye una variedad de operadores para realizar operaciones aritméticas, de comparación y lógicas.

FUNCIONES INTEGRADAS
Python proporciona muchas funciones integradas útiles, como int y float para convertir tipos de datos, entre otras.

ATRIBUTOS
Todos los objetos en Python tienen atributos. Por ejemplo, los números tienen atributos como is_integer() o bit_length().

BOOLEANS
Los booleanos en Python son subclases de enteros con solo dos valores posibles: True y False.

STRINGS
Las cadenas en Python son secuencias de caracteres que se pueden definir con comillas simples, comillas dobles o comillas triples. Se pueden manipular y acceder a través de métodos y operaciones.

OPERACIONES CON CADENAS
Python ofrece una variedad de operaciones para trabajar con cadenas, como concatenación, acceso a caracteres individuales y operaciones de división.

MÉTODOS
Las cadenas en Python tienen muchos métodos útiles, como upper(), lower() y title().

SLICING
La técnica de "slicing" en Python permite acceder a partes específicas de una cadena utilizando índices y rangos.

PLANTILLAS
Python ofrece la función format() para insertar valores variables en cadenas.

CONDICIONALES
Python utiliza las palabras clave if, elif y else para ejecutar código condicionalmente basado en expresiones booleanas.

BUCLES
Python tiene bucles for y while que permiten ejecutar un bloque de código repetidamente hasta que se cumple una condición.

```bash
PRIMER PROGRAMA

Para ejecutar un programa Python, simplemente crea un archivo .py con tu código y ejecútalo con Python.

print("Hola mundo")
