# Recoñecemento de caras

OpenCV ten funcións para buscar obxectos polas imaxes. Para iso é o CascadeClassifier

```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("imaxes/fotoia3.jpg")
imaxe_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(imaxe_color, cv2.COLOR_BGR2GRAY)

plt.imshow(imaxe_color)
plt.title('Imaxe1')
plt.show()
plt.imshow(gray)
plt.title('Imaxe en gris')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor = 1.2,
    minNeighbors = 5,
    minSize = (100,100), 
    )
print("Caras = ", len(faces))    
```

```python

destino = imaxe_color.copy()
for (x,y,w,h) in faces:
    cv2.rectangle(destino, (x,y), (x+h, y+h), (0, 255, 0), 2)

plt.imshow(destino)
plt.title('Caras detectadas')

plt.show()
```

Para explorar un pouco imos a facer unha función

```python
def buscarCaras(imaxe):
  face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

  faces = face_cascade.detectMultiScale(
      imaxe,
      scaleFactor = 1.2,
      minNeighbors = 5,
      minSize = (30,30), 
      )
  print("Caras = ", len(faces))
  destino = imaxe.copy()
  for (x,y,w,h) in faces:
      cv2.rectangle(destino, (x,y), (x+h, y+h), (0, 255, 0), 2)

  plt.imshow(destino)
  plt.title('Caras detectadas')

  plt.show()
```

```python
imaxePhill = cv2.imread("imaxes/phill.png")
imaxePhillColor = cv2.cvtColor(imaxePhill, cv2.COLOR_BGR2RGB)
buscarCaras(imaxePhillColor)
```

Funciona ben

```python
imaxeTodos = cv2.imread("imaxes/todos.png")
imaxeTodosColor = cv2.cvtColor(imaxeTodos, cv2.COLOR_BGR2RGB)
buscarCaras(imaxeTodosColor)
```
