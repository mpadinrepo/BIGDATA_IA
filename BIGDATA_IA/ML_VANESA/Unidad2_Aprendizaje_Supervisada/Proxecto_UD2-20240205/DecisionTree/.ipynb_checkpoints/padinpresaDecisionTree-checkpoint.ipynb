### División del conjunto de datos en entrenamiento y prueba:


```python
from sklearn.model_selection import train_test_split

# Características (X) y etiquetas (y) para el conjunto de entrenamiento
X_train = df_train.drop('y', axis=1)
y_train = df_train['y']

# Características (X) para el conjunto de prueba
X_test = df_test

# División del conjunto de datos de entrenamiento en subconjuntos de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Verificación de las dimensiones de los conjuntos creados
print("Dimensiones del conjunto de entrenamiento:", X_train.shape)
print("Dimensiones del conjunto de validación:", X_val.shape)
print("Dimensiones del conjunto de prueba:", X_test.shape)
```

### Codificación de las características categóricas utilizando One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

# Codificar características categóricas para el conjunto de entrenamiento
one_hot_encoder = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = one_hot_encoder.fit_transform(X_train[['job']])
# Codificar características categóricas para el conjunto de validación
X_val_encoded = one_hot_encoder.transform(X_val[['job']])
# Codificar características categóricas para el conjunto de prueba
X_test_encoded = one_hot_encoder.transform(X_test[['job']])
```

```python
print("Dimensiones del conjunto de entrenamiento codificado:", X_train_encoded.shape)
print("Contenido del conjunto de entrenamiento codificado:")
print(X_train_encoded)
print("Dimensiones del conjunto de validación codificado:", X_val_encoded.shape)
print("Contenido del conjunto de validación codificado:")
print(X_val_encoded)
```

### Estandarización de las características numéricas

```python
from sklearn.preprocessing import StandardScaler

# Inicializar el scaler
scaler = StandardScaler()

# Estandarizar las características numéricas para el conjunto de entrenamiento
X_train_scaled = scaler.fit_transform(X_train.select_dtypes(include='number'))

# Estandarizar las características numéricas para el conjunto de validación
X_val_scaled = scaler.transform(X_val.select_dtypes(include='number'))

# Estandarizar las características numéricas para el conjunto de prueba
X_test_scaled = scaler.transform(X_test.select_dtypes(include='number'))

# Mostrar las dimensiones de los conjuntos estandarizados
print("Dimensiones del conjunto de entrenamiento estandarizado:", X_train_scaled.shape)
print("Dimensiones del conjunto de validación estandarizado:", X_val_scaled.shape)
print("Dimensiones del conjunto de prueba estandarizado:", X_test_scaled.shape)

# Mostrar el contenido del conjunto de entrenamiento estandarizado
print("Contenido del conjunto de entrenamiento estandarizado:")
print(X_train_scaled)

# Mostrar el contenido del conjunto de validación estandarizado
print("\nContenido del conjunto de validación estandarizado:")
print(X_val_scaled)

# Mostrar el contenido del conjunto de prueba estandarizado
print("\nContenido del conjunto de prueba estandarizado:")
print(X_test_scaled)

```

# 4. Creación del modelo:
En esta etapa, creamos el modelo Logistic Regression utilizando la clase KNeighborsClassifier de scikit-learn y ajustamos el modelo a los datos de entrenamiento.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Crear el modelo de árbol de decisión
modelo_arbol = DecisionTreeClassifier(random_state=42)

# Entrenar el modelo con el conjunto de entrenamiento estandarizado
modelo_arbol.fit(X_train_scaled, y_train)

# Hacer predicciones sobre el conjunto de validación
predicciones_val = modelo_arbol.predict(X_val_scaled)

# Calcular la precisión del modelo en el conjunto de validación
precision_val = accuracy_score(y_val, predicciones_val)
print("Precisión del modelo en el conjunto de validación:", precision_val)
```

Se visualiza la clasificación realizada por el modelo mediante gráficos adecuados.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve

# Matriz de Confusión
cm = confusion_matrix(y_val, y_val_pred)
plt.figure(figsize=(18, 12))

plt.subplot(2, 3, 1)
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusión')

# Curva ROC
fpr, tpr, _ = roc_curve(y_val, y_val_pred)
plt.subplot(2, 3, 2)
plt.plot(fpr, tpr, color='blue', lw=2, label='Curva ROC')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')

# Comparación entre Clasificación Real y Predicción
plt.subplot(2, 3, 3)
plt.scatter(range(len(y_val)), y_val, label='Valores Reales', alpha=0.5)
plt.scatter(range(len(y_val)), y_val_pred, label='Predicciones', alpha=0.5)
plt.xlabel('Instancias')
plt.ylabel('Clase')
plt.title('Comparación entre Clasificación Real y Predicción')
plt.legend()

# Distribución de Probabilidades Predichas
plt.subplot(2, 3, 4)
sns.kdeplot(y_val, label='Valores Reales', fill=True)
sns.kdeplot(y_val_pred, label='Predicciones', fill=True)
plt.xlabel('Clase')
plt.ylabel('Densidad')
plt.title('Distribución de Probabilidades Predichas')
plt.legend()

# Gráfico de Ajuste de Calibración
plt.subplot(2, 3, 5)
plt.hist([y_val, y_val_pred], bins=20, label=['Valores Reales', 'Predicciones'], alpha=0.7)
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.title('Gráfico de Ajuste de Calibración')
plt.legend()

plt.tight_layout()
plt.show()
```

# 7.  Cálculo del Mejor Parámetro (si Aplica):

```python
from sklearn.model_selection import GridSearchCV

# Definir los hiperparámetros a buscar
parametros = {
    'max_depth': [3, 5, 7, 9, 11],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Crear el modelo base
modelo_arbol = DecisionTreeClassifier(random_state=42)

# Crear el objeto GridSearchCV
grid_search = GridSearchCV(estimator=modelo_arbol, param_grid=parametros, cv=10, scoring='accuracy')

# Realizar la búsqueda en la cuadrícula
grid_search.fit(X_train_scaled, y_train)

# Mostrar el mejor parámetro encontrado
print("Mejor parámetro encontrado:", grid_search.best_params_)

# Obtener el mejor modelo
mejor_modelo = grid_search.best_estimator_

# Evaluar el mejor modelo en el conjunto de validación
precision_mejor_modelo = mejor_modelo.score(X_val_scaled, y_val)
print("Precisión del mejor modelo en el conjunto de validación:", precision_mejor_modelo)
```

- **parametros**: Especifica la cuadrícula de hiperparámetros que se probarán.
- **GridSearchCV**: Realiza la búsqueda en la cuadrícula con validación cruzada.
- **cv=5**: Indica que se utilizará validación cruzada de 5 divisiones para evaluar el rendimiento de cada combinación de hiperparámetros.
- **scoring='accuracy'**: Especifica que se utilizará la precisión como métrica de evaluación.
- **fit**: Ajusta el objeto GridSearchCV a los datos de entrenamiento.
- **best_params_**: Devuelve los mejores hiperparámetros encontrados durante la búsqueda en la cuadrícula.
- **best_estimator_**: Devuelve el mejor modelo encontrado con los mejores hiperparámetros.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve

# Hacer predicciones utilizando el mejor modelo
y_val_pred_mejorado = mejor_modelo.predict(X_val_scaled)

# Matriz de Confusión
cm_mejorado = confusion_matrix(y_val, y_val_pred_mejorado)
plt.figure(figsize=(18, 12))

plt.subplot(2, 3, 1)
sns.heatmap(cm_mejorado, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusión')

# Curva ROC
fpr_mejorado, tpr_mejorado, _ = roc_curve(y_val, y_val_pred_mejorado)
plt.subplot(2, 3, 2)
plt.plot(fpr_mejorado, tpr_mejorado, color='blue', lw=2, label='Curva ROC')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')

# Comparación entre Clasificación Real y Predicción
plt.subplot(2, 3, 3)
plt.scatter(range(len(y_val)), y_val, label='Valores Reales', alpha=0.5)
plt.scatter(range(len(y_val)), y_val_pred_mejorado, label='Predicciones', alpha=0.5)
plt.xlabel('Instancias')
plt.ylabel('Clase')
plt.title('Comparación entre Clasificación Real y Predicción')
plt.legend()

# Distribución de Probabilidades Predichas
plt.subplot(2, 3, 4)
sns.kdeplot(y_val, label='Valores Reales', fill=True)
sns.kdeplot(y_val_pred_mejorado, label='Predicciones', fill=True)
plt.xlabel('Clase')
plt.ylabel('Densidad')
plt.title('Distribución de Probabilidades Predichas')
plt.legend()

# Gráfico de Ajuste de Calibración
plt.subplot(2, 3, 5)
plt.hist([y_val, y_val_pred_mejorado], bins=20, label=['Valores Reales', 'Predicciones'], alpha=0.7)
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.title('Gráfico de Ajuste de Calibración')
plt.legend()

plt.tight_layout()
plt.show()
```
