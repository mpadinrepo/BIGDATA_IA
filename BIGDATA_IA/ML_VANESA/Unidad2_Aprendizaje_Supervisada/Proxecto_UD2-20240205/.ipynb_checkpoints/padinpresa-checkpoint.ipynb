# Parte 1: Carga de Datos

```python
import pandas as pd

# Cargar los datos
dataframeTrain = pd.read_csv("credit-train.csv")
dataframeTest = pd.read_csv("credit-test.csv")
```

# Parte 2: Preprocesamiento de Datos

```python
from sklearn.preprocessing import StandardScaler

def preprocess_data(dataframe):
    # Convertir variables categóricas en variables dummy
    dataframe = pd.get_dummies(dataframe, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome'])
    
    # Normalizar las características numéricas
    scaler = StandardScaler()
    numeric_features = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp_var_rate', 'cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed']
    dataframe[numeric_features] = scaler.fit_transform(dataframe[numeric_features])
    
    return dataframe

# Preprocesamiento de los datos de entrenamiento y prueba
X_train = dataframeTrain.drop(columns=['y'])
y_train = dataframeTrain['y']
X_train = preprocess_data(X_train)
X_test = preprocess_data(dataframeTest)
```

# Parte 3: Procesamiento y Evaluación de Algoritmos

## Regresión Logística

### Fase 1: Definición del Pipeline


```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

logistic_pipeline = Pipeline([
    ('feature_selection', SelectFromModel(RandomForestClassifier())),
    ('classification', LogisticRegression(max_iter=110000))
])
```

### Fase 2: Definición de Parámetros para Búsqueda de Hiperparámetros


```python
logistic_parameters = {
    'feature_selection__estimator__n_estimators': [50, 100, 200],
    'feature_selection__threshold': ['mean', 'median', '1.25*mean']
}
```

### Fase 3: Búsqueda de Hiperparámetros


```python
from sklearn.model_selection import GridSearchCV

logistic_grid_search = GridSearchCV(logistic_pipeline, logistic_parameters, cv=5, n_jobs=-1, scoring='f1')
logistic_grid_search.fit(X_train, y_train)
```

### Fase 4: Evaluación del Mejor Modelo


```python
logistic_best_model = logistic_grid_search.best_estimator_
```

```python
def extract_metrics(classification_report):
    lines = classification_report.split('\n')
    metrics = {}
    for line in lines[2:-5]:
        parts = line.split()
        if len(parts) > 0:
            metric_name = parts[0]
            metric_values = [float(x) for x in parts[1:]]
            metrics[metric_name] = metric_values
    return metrics
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_performance(models, names):
    plt.figure(figsize=(10, 6))
    metrics = ['precision', 'recall', 'f1-score', 'accuracy']
    for model, name in zip(models, names):
        scores = [model[metric][0] for metric in metrics]
        sns.lineplot(x=metrics, y=scores, label=name, marker='o')
    plt.title('Performance of Different Models')
    plt.xlabel('Metrics')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def extract_metrics(classification_report_str):
    lines = classification_report_str.split('\n')
    metrics = lines[-2].split()  # Obtener la última línea del informe de clasificación que contiene las métricas
    values = lines[-1].split()  # Obtener la línea siguiente que contiene los valores de las métricas
    metrics_dict = {metric: float(value) for metric, value in zip(metrics[1:], values[1:])}
    return metrics_dict

```
