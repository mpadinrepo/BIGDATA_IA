### Eliminar valores atípicos utilizando el método de los cuantiles

```python
import matplotlib.pyplot as plt

# Función para identificar y sustituir valores atípicos utilizando el método de los cuantiles más convenientes
def sustituir_valores_atipicos(df, columna):
    # Calcular los cuartiles
    q1 = df[columna].quantile(0.25)
    q3 = df[columna].quantile(0.75)
    
    # Calcular el rango intercuartílico
    iqr = q3 - q1
    
    # Definir los límites para identificar valores atípicos
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    # Sustituir valores atípicos con los cuartiles más convenientes
    df[columna] = df[columna].apply(lambda x: q1 if x < limite_inferior else q3 if x > limite_superior else x)
    
    return df



```

#### Análisis del Informe de Clasificación:

El informe de clasificación proporciona métricas detalladas sobre el rendimiento del modelo de clasificación en cada clase y globalmente. Aquí hay una interpretación de las métricas más importantes:

- **Precision**: Es la proporción de verdaderos positivos (predicciones correctas) sobre todos los positivos predichos por el modelo. Para la clase 0, la precisión es alta (0.90), lo que indica que la mayoría de las muestras clasificadas como clase 0 son realmente de esa clase. Sin embargo, para la clase 1, la precisión es mucho más baja (0.53), lo que indica que el modelo tiende a clasificar algunas muestras incorrectamente como clase 1.

- **Recall**: Es la proporción de verdaderos positivos (predicciones correctas) sobre todos los ejemplos que pertenecen realmente a la clase. Para la clase 0, el recall es alto (0.98), lo que indica que el modelo identifica la gran mayoría de las muestras que realmente son de clase 0. Sin embargo, para la clase 1, el recall es bajo (0.19), lo que indica que el modelo pierde una gran cantidad de muestras que realmente son de clase 1.

- **F1-score**: Es una medida que combina precision y recall en una sola métrica. Es útil cuando hay un desequilibrio entre las clases. El F1-score se calcula como la media armónica de precision y recall. Para la clase 0, el F1-score es alto (0.94), lo que indica un buen equilibrio entre precision y recall. Sin embargo, para la clase 1, el F1-score es bajo (0.28), lo que sugiere que el modelo tiene dificultades para clasificar correctamente las muestras de esta clase.

- **Accuracy**: Es la proporción de predicciones correctas (verdaderos positivos y verdaderos negativos) sobre el total de muestras. La precisión global del modelo es del 89%, lo que indica que el modelo clasifica correctamente el 89% de todas las muestras.

- **Macro avg y weighted avg**: Estas son las medias de las métricas para todas las clases. La macro avg calcula la media no ponderada de precision, recall y F1-score para todas las clases, mientras que la weighted avg calcula la media ponderada por el soporte (número de muestras verdaderas de cada clase).

### Conclusiones:

- El modelo muestra un buen rendimiento en la clasificación de la clase mayoritaria (clase 0), con alta precision, recall y F1-score.
- Sin embargo, el rendimiento en la clasificación de la clase minoritaria (clase 1) es deficiente, con baja precision, recall y F1-score.
- Esto sugiere que el modelo tiene dificultades para generalizar patrones para la clase minoritaria y tiende a clasificarla incorrectamente como la clase mayoritaria.
- Para mejorar el rendimiento del modelo, podrían explorarse técnicas como el ajuste de umbrales de clasificación, el uso de algoritmos de clasificación más avanzados o el tratamiento del desequilibrio de clases mediante técnicas como el muestreo de datos o el ajuste de pesos de las clases.
