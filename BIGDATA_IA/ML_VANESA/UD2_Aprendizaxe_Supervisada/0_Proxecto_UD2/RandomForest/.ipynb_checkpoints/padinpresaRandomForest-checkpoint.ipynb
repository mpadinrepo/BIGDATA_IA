### División del conjunto de datos en entrenamiento , prueba y validación:

```python
from sklearn.model_selection import train_test_split

# Definir X_train y y_train para df_train
X_train = df_train.drop(columns=['y'])  # Características
y_train = df_train['y']  # Variable objetivo

# Dividir el conjunto de entrenamiento en entrenamiento y validación
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Mostrar las dimensiones de los conjuntos divididos
print("Dimensiones del conjunto de entrenamiento:", X_train_split.shape)
print("Dimensiones del conjunto de validación:", X_val_split.shape)
print("Dimensiones del conjunto de prueba:", df_test.shape)

```

```python
# Mostrar las primeras filas de los conjuntos de datos
print("\nPrimeras filas del conjunto de entrenamiento:")
print(X_train.head())

print("\nPrimeras filas del conjunto de validación:")
print(X_val_split.head())

print("\nPrimeras filas del conjunto de prueba:")
print(X_test.head())
```

### Codificación de las características categóricas utilizando One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Definir columnas categóricas y numéricas
categorical_cols = X_train.select_dtypes(include=['object']).columns
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns

# Definir transformadores para columnas categóricas y numéricas
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Combinar transformadores en un preprocesador
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numeric_transformer, numeric_cols)
    ])

# Codificar las variables en el conjunto de entrenamiento
X_train_encoded = preprocessor.fit_transform(X_train_split)

# Codificar las variables en el conjunto de validación
X_val_encoded = preprocessor.transform(X_val_split)

# Codificar las variables en el conjunto de prueba
X_test_encoded = preprocessor.transform(X_test)
```

# 4. Creación del modelo:
En esta etapa, creamos el modelo Random Forest utilizando RandomForestClassifier de scikit-learn y ajustamos el modelo a los datos de entrenamiento.

***En este código:*** 

Importamos la clase RandomForestClassifier de sklearn.ensemble, que es el modelo que utilizaremos.
Inicializamos un clasificador de Random Forest con un estado aleatorio (random_state) establecido en 42 para garantizar la reproducibilidad de los resultad
os.
Entrenamos el modelo utilizando el conjunto de entrenamiento dividido y codificado (X_train_encoded y y_train_spl
it).
Realizamos predicciones en el conjunto de validación codificado (X_val_encoded) utilizando el método pre
dict.
Realizamos predicciones en el conjunto de prueba codificado (X_test_encoded) utilizando el mismo modelo entr
enado.
Este proceso nos permite evaluar el rendimiento del modelo en el conjunto de validación y obtener predicciones para el conjunto de prueba.

```python
from sklearn.ensemble import RandomForestClassifier

# Inicializar el clasificador de Random Forest
random_forest = RandomForestClassifier(random_state=42)

# Entrenar el modelo con el conjunto de entrenamiento dividido y codificado
random_forest.fit(X_train_encoded, y_train_split)

# Realizar predicciones en el conjunto de validación codificado
predictions_val = random_forest.predict(X_val_encoded)

# Realizar predicciones en el conjunto de prueba codificado
predictions_test = random_forest.predict(X_test_encoded)
```

# 5. Análisis de los Resultados
Realizamos un análisis de los resultados obtenidos por el modelo Random Forest, como la precisión y otras métricas de evaluación del rendimiento del modelo.

```python
from sklearn.metrics import classification_report
# Definir las etiquetas verdaderas del conjunto de prueba
y_test = df_test['y']

# Realizar predicciones en el conjunto de prueba
y_pred_test = random_forest.predict(X_test_encoded)

# Definir las etiquetas verdaderas del conjunto de prueba
y_test = df_test['y']

# Reporte de clasificación para el conjunto de prueba
classification_rep_test = classification_report(y_test, y_pred_test)

print("Reporte de Clasificación para el conjunto de prueba:\n", classification_rep_test)

```

# 6. Gráficos de la clasificación obtenida: 
Se visualiza la clasificación realizada por el modelo mediante gráficos adecuados.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve

# Matriz de Confusión
plt.figure(figsize=(18, 12))
plt.subplot(2, 3, 1)
sns.heatmap(confusion_matrix(y_val_split, predictions_val), annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusión para el conjunto de validación')

# Curva ROC
fpr, tpr, _ = roc_curve(y_val_split, predictions_val)
plt.subplot(2, 3, 2)
plt.plot(fpr, tpr, color='blue', lw=2, label='Curva ROC')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC para el conjunto de validación')
plt.legend(loc='lower right')

# Comparación entre Clasificación Real y Predicción
plt.subplot(2, 3, 3)
plt.scatter(range(len(y_val_split)), y_val_split, label='Valores Reales', alpha=0.5)
plt.scatter(range(len(y_val_split)), predictions_val, label='Predicciones', alpha=0.5)
plt.xlabel('Instancias')
plt.ylabel('Clase')
plt.title('Comparación entre Clasificación Real y Predicción para el conjunto de validación')
plt.legend()

# Distribución de Probabilidades Predichas
plt.subplot(2, 3, 4)
sns.kdeplot(y_val_split, label='Valores Reales', fill=True)
sns.kdeplot(predictions_val, label='Predicciones', fill=True)
plt.xlabel('Clase')
plt.ylabel('Densidad')
plt.title('Distribución de Probabilidades Predichas para el conjunto de validación')
plt.legend()

# Gráfico de Ajuste de Calibración
plt.subplot(2, 3, 5)
plt.hist([y_val_split, predictions_val], bins=20, label=['Valores Reales', 'Predicciones'], alpha=0.7)
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.title('Gráfico de Ajuste de Calibración para el conjunto de validación')
plt.legend()

plt.tight_layout()
plt.show()

```

# 7.  Cálculo del Mejor Parámetro (si Aplica):

```python
from sklearn.model_selection import GridSearchCV

# Definir los hiperparámetros que deseas ajustar
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Inicializar el clasificador RandomForestClassifier
random_forest = RandomForestClassifier(random_state=2024)

# Inicializar la búsqueda en cuadrícula con validación cruzada
grid_search = GridSearchCV(estimator=random_forest, param_grid=param_grid, cv=5, scoring='accuracy')

# Realizar la búsqueda en cuadrícula en los datos de entrenamiento
grid_search.fit(X_train_split, y_train_split)

# Mostrar los mejores hiperparámetros encontrados
print("Mejores hiperparámetros encontrados:")
print(grid_search.best_params_)

# Mostrar la mejor puntuación obtenida durante la búsqueda
print("Mejor puntuación durante la búsqueda:", grid_search.best_score_)
```

- **parametros**: Especifica la cuadrícula de hiperparámetros que se probarán.
- **GridSearchCV**: Realiza la búsqueda en la cuadrícula con validación cruzada.
- **cv=5**: Indica que se utilizará validación cruzada de 5 divisiones para evaluar el rendimiento de cada combinación de hiperparámetros.
- **scoring='accuracy'**: Especifica que se utilizará la precisión como métrica de evaluación.
- **fit**: Ajusta el objeto GridSearchCV a los datos de entrenamiento.
- **best_params_**: Devuelve los mejores hiperparámetros encontrados durante la búsqueda en la cuadrícula.
- **best_estimator_**: Devuelve el mejor modelo encontrado con los mejores hiperparáms.stos.
