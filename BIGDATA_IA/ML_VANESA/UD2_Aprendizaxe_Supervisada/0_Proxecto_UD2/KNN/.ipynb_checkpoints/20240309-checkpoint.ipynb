### Eliminar valores atípicos utilizando el método de los cuantiles

```python
import matplotlib.pyplot as plt

# Función para identificar y sustituir valores atípicos utilizando el método de los cuantiles más convenientes
def sustituir_valores_atipicos(df, columna):
    # Calcular los cuartiles
    q1 = df[columna].quantile(0.25)
    q3 = df[columna].quantile(0.75)
    
    # Calcular el rango intercuartílico
    iqr = q3 - q1
    
    # Definir los límites para identificar valores atípicos
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    # Sustituir valores atípicos con los cuartiles más convenientes
    df[columna] = df[columna].apply(lambda x: q1 if x < limite_inferior else q3 if x > limite_superior else x)
    
    return df



```

```python
# Grafico diagrama de caja y bigotes para la columna 'age' antes de sustituir valores atípicos
plt.figure(figsize=(8,6))
plt.boxplot(df_train['age'])
plt.title('Diagrama de Caja y Bigotes - Age (Antes)')
plt.ylabel('Edad')
plt.show()

# Sustituir valores atípicos en la columna 'age'
df_train_sin_atipicos = sustituir_valores_atipicos(df_train.copy(), 'age')

# Grafico diagrama de caja y bigotes para la columna 'age' después de sustituir valores atípicos
plt.figure(figsize=(8,6))
plt.boxplot(df_train_sin_atipicos['age'])
plt.title('Diagrama de Caja y Bigotes - Age (Después)')
plt.ylabel('Edad')
plt.show()
```

# 4. Creación del modelo:
En esta etapa, creamos el modelo KNN utilizando la clase KNeighborsClassifier de scikit-learn y ajustamos el modelo a los datos de entrenamiento.

```python
from sklearn.neighbors import KNeighborsClassifier

# Definir el clasificador KNN con k=5
model = KNeighborsClassifier(n_neighbors=5)

# Entrenar el modelo
model.fit(X_train, y_train)
```

# 5. Análisis de los Resultados
Realizamos un análisis de los resultados obtenidos por el modelo KNN, como la precisión y otras métricas de evaluación del rendimiento del modelo.

```python
from sklearn.metrics import classification_report

# Realizar predicciones
y_pred = model.predict(X_test)

# Reporte de clasificación
classification_rep = classification_report(y_test, y_pred)
print("Reporte de Clasificación:\n", classification_rep)

```

# 6. Gráficos de la clasificación obtenida:

```python
# Paso 8: Gráficos de la Clasificación Obtenida

# Importar librerías necesarias
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión')
plt.xlabel('Etiqueta Predicha')
plt.ylabel('Etiqueta Real')
plt.show()

# Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Curva ROC (área = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Real')
plt.scatter(range(len(y_pred)), y_pred, color='red', marker='x', label='Predicción')
plt.title('Comparación entre Clasificación Real y Predicción')
plt.xlabel('Instancias')
plt.ylabel('Clase')
plt.legend()
plt.show()
# Obtener probabilidades predichas para la clase 1
proba_pred_class1 = model.predict_proba(X_test)[:, 1]

# Graficar la densidad de probabilidad
plt.figure(figsize=(8, 6))
sns.kdeplot(proba_pred_class1[y_test == 0], color='blue', shade=True, label='Clase 0')
sns.kdeplot(proba_pred_class1[y_test == 1], color='red', shade=True, label='Clase 1')
plt.title('Distribución de Probabilidades Predichas')
plt.xlabel('Probabilidad de Clase 1')
plt.ylabel('Densidad')
plt.legend()
plt.show()
from sklearn.calibration import calibration_curve

# Obtener la curva de calibración
prob_true, prob_pred = calibration_curve(y_test, proba_pred_class1, n_bins=10)

# Graficar la curva de calibración
plt.figure(figsize=(8, 6))
plt.plot(prob_pred, prob_true, marker='o', color='blue', label='Curva de Calibración')
plt.plot([0, 1], [0, 1], linestyle='--', color='red', label='Línea Diagonal')
plt.title('Gráfico de Ajuste de Calibración')
plt.xlabel('Probabilidad Predicha')
plt.ylabel('Frecuencia Real')
plt.legend()
plt.show()

```
